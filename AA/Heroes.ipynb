## Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#features como Alingment e Gender deixam o modelo muito enviesado

#Mapping cat data
#heroes['Alignment'] = heroes['Alignment'].map({'Good': 0, "Neutral": 1, "Bad":2})
#heroes['Gender'] = heroes['Gender'].map({"Male": 0, "Female":1})

features = ['Intelligence', 'Strength', 'Speed', 'Durability', 'Power', 'Combat']
```

```python
#Normalizando features (Gauss)

st_heroes = heroes.copy()
scaler = StandardScaler()

st_heroes[features] = scaler.fit_transform(heroes[features])
st_heroes
```

```python
#iteraÃ§Ã£o para diferentes nÃºmeros de clusters

inertia = []
for i in range(1, 5):
    kmeans = KMeans(i, random_state=42)
    kmeans.fit_predict(st_heroes[features])
    inertia.append(kmeans.inertia_)
```

```python
#grÃ¡fico de inertia x nÃºmero de clusters

plt.plot(range(1, 5) , inertia, 'bx-')
plt.xlabel('i')
plt.ylabel('Inertia')
plt.show()
```

```python
st_heroes['clusters'] = kmeans.labels_
```

```python
st_heroes.head()
```

```python
pd.Series(kmeans.labels_).unique()
```

```python
#Estamos trabalhando com um espaÃ§o n-dimensional, onde n > 3, geomÃ©tricamente inviÃ¡vel
#Reduzir dimensionalidade para n = 2, permitindo a visualizaÃ§Ã£o

from sklearn.decomposition import PCA
pca = PCA(n_components=2, whiten=True, random_state=42)
sliced = st_heroes[features]
sliced['clusters'] = kmeans.labels_

sliced_pca = pca.fit_transform(sliced)
df_clusters = pd.DataFrame(data=sliced_pca, columns=['x', 'y'])
df_clusters['clusters'] = kmeans.labels_
```

```python
df_clusters['clusters'] = df_clusters['clusters'].map({0: 'Realistas', 1:'Lutadores', 2:'Estrategistas', 3:'Overpowered'})
df_clusters.head()
```

```python
plt.figure(figsize=(12,8))
palette = sns.color_palette("bright", 4)
sns.scatterplot(x='x', y='y', hue='clusters', data=df_clusters, 
                palette=palette, legend='full')
plt.legend()
plt.xlim(-3,3)
plt.ylim(-4,4)
plt.show()
```
